```python
import pandas as pd

X_train = pd.read_csv('X_train.csv')
X_test = pd.read_csv('X_test.csv')
y_train = pd.read_csv('y_train.csv')
y_test = pd.read_csv('y_test.csv')
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, VotingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import GridSearchCV
```

```python
rf = RandomForestClassifier(random_state=42)
xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
svm = SVC(probability=True, random_state=42)
knn = KNeighborsClassifier()
```

```python
ensemble_model = VotingClassifier(
    estimators=[
        ('rf', rf),
        ('xgb', xgb),
        ('svm', svm),
        ('knn', knn)
    ],
    voting='soft'  # hard=majority voting and soft=average of probabilities
)
```

```python
'''taking sample due to hardware limitation,i cant not train on whole dataset...because my laptop will die!!!!i tried once to train on
whole dataset got my laptop crash sucessfully, then realized how big the dataset is....so my dear teammates dont ask y!!!! and just move on'''

X_train_sampled = X_train.sample(frac=0.01, random_state=42)
y_train_sampled = y_train.loc[X_train_sampled.index]
```

```python
X_train_sampled = X_train_sampled.dropna()
y_train_sampled = y_train_sampled.loc[X_train_sampled.index]
```

```python
#ensemble_model.fit(X_train, y_train) #commanted out because of hardware limitation
ensemble_model.fit(X_train_sampled, y_train_sampled)
```

```python
X_test_cleaned = X_test.dropna()
y_test_cleaned = y_test.loc[X_test_cleaned.index]
```

```python
X_test_filled = X_test.fillna(X_test.mean())
```

```python
y_pred = ensemble_model.predict(X_test_cleaned)
```

```python
accuracy = accuracy_score(y_test_cleaned, y_pred)
print(f"Accuracy: {accuracy:.2f}")

report = classification_report(y_test_cleaned, y_pred)
print("Classification Report:\n", report)
```

```python
param_grid = {
    'n_estimators': [50, 100, 200, 300, 500],
    'max_depth': [None, 10, 20, 30, 40],
    'min_samples_split': [2, 5, 10, 15],
    'min_samples_leaf': [1, 2, 4, 6],
    'max_features': ['auto', 'sqrt', 'log2'],
}

grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, scoring='accuracy', cv=3, n_jobs=-1, verbose=2)
grid_search.fit(X_train_sampled, y_train_sampled)

print("Best Parameters:", grid_search.best_params_)
```

```python
import joblib
joblib.dump(ensemble_model, 'ensemble_model.pkl')
```

```python
best_rf = grid_search.best_estimator_
y_pred = best_rf.predict(X_test)

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

report = classification_report(y_test, y_pred)
print("Classification Report:\n", report)

conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", conf_matrix)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Class 0', 'Class 1'], 
            yticklabels=['Class 0', 'Class 1'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()
```
