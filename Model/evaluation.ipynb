```python
import pandas as pd
import joblib
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import json
```

```python
X_test = pd.read_csv('X_test.csv')
y_test = pd.read_csv('y_test.csv')
```

```python
X_test_cleaned = X_test.dropna()
y_test_cleaned = y_test.loc[X_test_cleaned.index] 
```

```python
ensemble_model = joblib.load('ensemble_model.pkl')
```

```python
y_pred = ensemble_model.predict(X_test_cleaned)
```

Accuracy

```python
accuracy = accuracy_score(y_test_cleaned, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

classification report

```python
class_report = classification_report(y_test_cleaned, y_pred, output_dict=True)
print("Classification Report:\n", class_report)
```

confusion matrix

```python
conf_matrix = confusion_matrix(y_test_cleaned, y_pred)
print("Confusion Matrix:\n", conf_matrix)
```

```python
evaluation_results = {
    'accuracy': accuracy,
    'classification_report': class_report,
    'confusion_matrix': conf_matrix.tolist()
}
with open('evaluation_results.json', 'w') as f:
    json.dump(evaluation_results, f)

print("Evaluation results saved to evaluation_results.json")
```
